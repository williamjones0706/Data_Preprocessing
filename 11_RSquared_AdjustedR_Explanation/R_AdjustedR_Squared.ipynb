# R and Adjusted R-Squared Explanation

# R-Squared

R2 is a measure of how well the regression model/equation explains the data or is fitted to the data relative to a simle average line. The Sum of Square residuals, noted above as SSE, is based on the ordinary least squares (OLS) method, where the model minimizes the sum of squared errors from y actual to y predicted values. This is divided by the sum of squared error total SST, where y actual error is measured against y average. By minizing the error in the Sum of Squared residuals, the R2 increases, better representing the data.

# Adjusted R-Squared

R Squared works well for single variable regression, but fails to accurately capture the fit of multivariable models. As you continue to add more variables, R squared can never decrease because SST will remain the same while SSE will increase or stay the same for R squared. So you get a fraction whose numerator only stays the same or increases (1- SSE/SST) with more variables.

The (1-R2) portion punishes the model for adding new variables that are not helpful. In other words the R2 would be low, making 1-R2 a high number like .98
On the other hand the n-1/n-p-1 is rewarded for adding new variables given that p=number of independent variables. So there is a balance that increases the adjusted R2 when new variables are included for multivariable analysis but only if they are useful in fitting the model.

This is why adjusted R squared is used for explaining the fit for multivariable models. If allows you to determine if adding new variables supported a better fit model.
