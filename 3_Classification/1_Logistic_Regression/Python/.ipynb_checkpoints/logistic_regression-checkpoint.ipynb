# Problem Scenario
We are data scientists working at a car company. The company want to predict whether or not customers will buy a certain SUV based on past customer age and salary data for buying similar SUVs. The results of the regression model will be used by ad teams to target ads for areas where the target customers reside.

## Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv("Social_Network_Ads.csv")
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

# I checked to make sure the X and y matrix/vector were read properly before moving on
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)

# make sure to check the outputs by printing before moving on
# usually recommend .2 for test set but because there are 400 customers in the data a .25 gives a nice rounded 300 and 100(test set)
```

```python
X_train[:10, :]
```

```python
y_train
```

```python
X_test[:10,:]
```

```python
y_test
```

## Feature Scaling

```python
# while some model type likes SVR require feature scaling, others are optimized by it but dont require it.
# this model type doesnt require it but it will help optimize the model, especially given the varying salary range compared to age


from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
# scale all features, dependent variable doesnt need it, already in scale 0/1
# in some cases we only scale certain features, in which case you'll use X_train[:,_whatever columns you need] =
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train[:20,:]

# We can see that now both the test and training set have each feature in the same range.
# Feature scaling usually results in a range that goes from -3 to 3
```

```python
X_test[:20,:]
```

## Training the Logistic Regression model on the Training set

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)

# in a different file we will include more parameter tuning
```

## Predicting a new result

```python
classifier.predict(sc.transform([[30, 87000]]))
# need to scale the input since the model was trained on scaled features
# you can also use predict_proba to get the probability as opposed to outcome prediction
# in this case according to the test set first customer result the model guessed correctly
```

## Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
np.set_printoptions(precision=2)
# this np.set above can be removed for this one since we dont need to force a number of decimals to show
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test), 1)), 1))

# the code above concatenates the pred and test results for easy visual comparison in a 2 d matrix
# the dependent vartiable vectors are reshaped to [len(results), 1] in order to do so
```

## Making the Confusion Matrix

```python
# the confusion matrix will allow us to see exactly how many correct/incorrect predictions were made and % accuracy_score
# confusion matrices are read with the prediction on top columns and true values in rows
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

## Visualising the Training set results

```python
# please do not re create this code by hand, it is complex
# basically we create a grid that is extremely dense and stepped by 0.25 for each x and y scale
# The scales are also inversed to make sense of the scaling that was done
# the final graph shows the line where our model will predict who buys and who doesnt (1, 0) 
# and layers the training predictions on top

from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_train), y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min()-10, stop = X_set[:, 0].max()+10, step = 0.25), 
                     np.arange(start = X_set[:, 1].min()-1000, stop = X_set[:, 1].max()+1000, step = 0.25))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape), 
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Training Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```python
# please do not re create this code by hand, it is complex
# basically we create a grid that is extremely dense and stepped by 0.25 for each x and y scale
# The scales are also inversed to make sense of the scaling that was done
# the final graph shows the line where our model will predict who buys and who doesnt (1, 0) 
# and layers the test set based predictions on top

from matplotlib.colors import ListedColormap

X_set_test, y_set_test = sc.inverse_transform(X_test), y_test
X1_test, X2_test = np.meshgrid(np.arange(start = X_set_test[:, 0].min()-10, stop = X_set_test[:, 0].max()+10, step = 0.25), 
                     np.arange(start = X_set_test[:, 1].min()-1000, stop = X_set_test[:, 1].max()+1000, step = 0.25))
plt.contourf(X1_test, X2_test, classifier.predict(sc.transform(np.array([X1_test.ravel(), X2_test.ravel()]).T)).reshape(X1_test.shape), 
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1_test.min(), X1_test.max())
plt.ylim(X2_test.min(), X2_test.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set_test[y_set_test == j, 0], X_set_test[y_set_test == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test Set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
