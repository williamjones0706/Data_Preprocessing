# Random Forest Regression

## Random forest explanation
The random forest model type is a type of ensemble learning model that essentialy uses N decision trees to predict y values form all the different trees and then average the values for final predictions. By default usually these algorithms are using around ~500 decision trees, hence the forest.

AN exmaple of ensemble learning, not perfectly related to random forest but useful, would be taking of average or median of 1000 people's guesses at how many coins are in a jar. It is akin to the central limit theorum, where the more observations you have the more likely you are to predict closer to the true mean

## Importing the libraries

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset

```
dataset = pd.read_csv("Position_Salaries.csv")
# Use 1:-1 for the column refernce in iloc because only using 1 will not yield a 2-d matrix
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

## Training the Random Forest Regression model on the whole dataset

```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X, y)
```

## Predicting a new result

```
regressor.predict([[6.5]])
```

## Visualising the Random Forest Regression results (higher resolution)

```
# As discussed in other sections, this example only includes a 2-d dataset to make it easier to visua
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape(len(X_grid), 1)
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, regressor.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Random Forest Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary (millions)')
plt.show()
```
